```python
import csv
import os

input_folder = "/Users/user/Downloads/Algotrade/Algotrade-forex/backtester_oop/backtester/data/datasets" 
output_folder = "/Users/user/Downloads/Algotrade/Algotrade-forex/backtester_oop/backtester/data/datasets/data_clean"

# Buat folder output bila belum ada
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.lower().endswith(".csv"):
        input_file = os.path.join(input_folder, filename)
        output_file = os.path.join(
            output_folder, 
            filename.replace(".csv", "_clean.csv")
        )

        with open(input_file, "r") as f_in, open(output_file, "w", newline="") as f_out:
            reader = csv.reader(f_in)
            writer = csv.writer(f_out)

            first_row = True
            for row in reader:
                cleaned = row[0].replace('"', '').replace('<', '').replace('>', '')
                cols = cleaned.split('\t')

                # Ubah header jadi lowercase
                if first_row:
                    cols = [col.lower() for col in cols]
                    first_row = False

                writer.writerow(cols)

        print(f"Converted: {filename} → {output_file}")

print("Selesai batch convert semua CSV!")
```
